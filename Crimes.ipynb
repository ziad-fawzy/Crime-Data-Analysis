```python
df['Weapon Desc'].fillna('Unknown',inplace=True)
df['Vict Sex'].fillna('Unknown',inplace=True)
df['Vict Descent'].fillna('Unknown',inplace=True)
```

# 1- Peak Crime Hour

```python
df['Hour']=df['TIME OCC']//100
```

```python
# حساب عدد الجرائم في كل ساعة
hour_counts=df['Hour'].value_counts().sort_index()
# استخراج ساعة الذروة
peak_crime_hour=int(hour_counts.idxmax())
print(f'The peak crime hour is: {peak_crime_hour}')
fig1=px.line(x=hour_counts.index,y=hour_counts.values,title='Crimes by Hour of Day',labels={'x':'Hour','y':'Number of Crimes'},markers=True)
fig1.show()
```

```python
### Bonus ###
# 1. تحويل عمود التاريخ إلى datetime (تأكد من مطابقة صيغة التاريخ في ملفك)
df['Date_Cleaned'] = pd.to_datetime(df['DATE OCC'])

# 2. استخراج اسم يوم الأسبوع
df['Day_of_Week'] = df['Date_Cleaned'].dt.day_name()

# 3. حساب عدد الجرائم لكل يوم وترتيبها أيام الأسبوع منطقياً
days_order = ['Saturday', 'Sunday','Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']
weekly_crimes = df['Day_of_Week'].value_counts().reindex(days_order).reset_index()
weekly_crimes.columns = ['Day of Week', 'Crime Count']

# --- رسم بياني خطي تفاعلي (Line Chart) يوضح الحركة عبر الأيام ---
fig5= px.line(weekly_crimes, x='Day of Week', y='Crime Count', text='Crime Count',
                     title='Crime Trend Across Days of the Week',
                     labels={'Day of Week': 'Day', 'Crime Count': 'Number of Crimes'},
                     markers=True)
fig5.update_traces(textposition="top center")
fig5.show()
```

```python
### Bonus ###
df['Month'] = pd.to_datetime(df['DATE OCC']).dt.month

month_counts = df.groupby('Month').size().reset_index(name='Count')

fig6 = px.line(month_counts,x='Month',y='Count',title='Crimes by Month',markers=True)
fig6.show()
```

```python
# تصفية البيانات للفترة الليلية (من 10 مساءً إلى 3:59 صباحاً)
night=df[(df['Hour']>=22) | (df['Hour']<=3)]
# عدد الجرائم لكل منطقة
# اعلي 10 مناطق
night_counts=(night.groupby('AREA NAME').size().reset_index(name='Count').sort_values('Count',ascending=False).head(10))
print(f'The area with the most night crimes is: {night_area}')
fig2=px.bar(night_counts.sort_values('Count'),x='Count',y='AREA NAME',title='Top 10 Areas-Night Crimes',labels={'AREA NAME':'Area','Count':'Number of Crimes'})
fig2.show()
```

```python
### Bonus ###
# حساب أعلى 5 مناطق تكررت فيها الجرائم في البيانات كاملة
top_areas = df['AREA NAME'].value_counts().head(5).reset_index()
top_areas.columns = ['Area Name', 'Total Crimes']

# --- رسم بياني شريطي تفاعلي عمودي ---
fig7 = px.bar(top_areas, x='Area Name', y='Total Crimes',title='Top 5 Most Dangerous Areas Overall',labels={'Area Name': 'Area', 'Total Crimes': 'Number of Crimes'})
fig7.show()
```

# 3- Victim Demographics

```python
# تصفية الأعمار واستبعاد القيم غير المنطقية (0 و 99)
df_age=df[(df['Vict Age']>0) & (df['Vict Age']<99)]
# تحديد حدود الفئات العمرية والأسماء المطلوبة
bins=[0,17,25,34,44,54,64,120]
labels=['0-17','18-25','26-34','35-44','45-54','55-64','65+']
# تقسيم الأعمار وحساب التكرار وترتيبها بالتتابع الصحيح
df_age['Age Group']=pd.cut(df_age['Vict Age'],bins=bins,labels=labels)
victim_ages=df_age['Age Group'].value_counts().reindex(labels)
# --- الرسم البياني التفاعلي ---
fig3=px.bar(victim_ages,x=victim_ages.index,y=victim_ages.values,title='Victim Age Distribution',labels={'x':'Age Group','y':'Number of Victims'})
fig3.show()
```

# 4- Anomaly Detection

```python
# سنستخدم العمود الأصلي كامل 'Vict Age' لرصد الشذوذ بالرسم البياني
# حساب عدد القيم الشاذة إحصائياً بالبايثون (طريقة IQR)
q1=df['Vict Age'].quantile(0.25)
q3=df['Vict Age'].quantile(0.75)
iqr=q3-q1
lower=q1-1.5*iqr
upper=q3+1.5*iqr
outliers=df[(df['Vict Age']<lower) | (df['Vict Age']>upper)]
print(f'The number of outliers is: {len(outliers)}')
print(f'the lower age is: {lower}')
print(f'the upper age is: {upper}')
# --- الرسم البياني التفاعلي (Box Plot) ---
fig4=px.box(df[df['Vict Age']>0],y='Vict Age',title='Outliers in Victim Age')
fig4.show()
```

```python
### Bonus ###
fig8=px.histogram(df[df['Vict Age']>0],x='Vict Age',nbins=30,title='Victim Age Distribution')
fig8.show()
```
